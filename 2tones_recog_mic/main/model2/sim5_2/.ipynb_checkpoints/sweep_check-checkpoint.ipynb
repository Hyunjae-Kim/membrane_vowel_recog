```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils as utils
import torch.nn.init as init
from torchvision import datasets
```

```python
model_num = 2
data_num = 5
point_len = 5
ch_num = 19
run_type='sweep2'
time_len =  point_len/10
fc_len = int(point_len/32)+1
```

```python
class Model(nn.Module):
    def __init__(self, loss):
        super(Model, self).__init__()
        input_c = ch_num
        channel = 64
        self.conv1 = nn.Sequential(
            nn.Conv1d(input_c, channel, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(channel))
        self.conv2 = nn.Sequential(
            nn.Conv1d(channel, channel*2, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(channel*2))
        self.conv3 = nn.Sequential(
            nn.Conv1d(channel*2, channel*2, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(channel*2))
        self.conv4 = nn.Sequential(
            nn.Conv1d(channel*2, channel*4, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(channel*4))
        self.conv5 = nn.Sequential(
            nn.Conv1d(channel*4, channel*4, kernel_size=5, stride=2, padding=2),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(channel*4))
        self.fc1 = nn.Sequential(
            nn.Linear(channel*4*fc_len, 1024),
            nn.LeakyReLU(negative_slope=0.1),
            nn.BatchNorm1d(1024))
        self.fc2 = nn.Sequential(
            nn.Linear(1024, 20),
            nn.Softmax(dim=1))
        self.loss = loss
        
    def forward(self, data, target):
        x = self.conv1(data)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.conv5(x).contiguous()
        x = x.view(x.size()[0],-1)
        x = self.fc1(x)
        h = self.fc2(x)
        
        l = self.loss(h, target)
        return l, h, target
```

```python
def test(mod_te, teX, teY, bat_te, dev, te_loss, te_acc):
    mod_te.eval()
    dloss_te = 0
    dacc_te = 0
    
    for i in range(bat_te[0]):
        loss_te, output_te, target_te = mod_te(teX[i*bat_te[1]:(i+1)*bat_te[1]], 
                                              teY[i*bat_te[1]:(i+1)*bat_te[1]])
        loss_te = loss_te.sum()
        
        _, output_te = torch.topk(output_te, 2)
        output_te, _ = output_te.sort()
        _, target_te = torch.topk(target_te, 2)
        target_te, _ = target_te.sort()

        dloss_te += loss_te.cpu().item()
        dacc_te += torch.sum(torch.sum(output_te==target_te, 1)==2).item()
        
    te_loss.append(dloss_te/bat_te[0])
    te_acc.append(dacc_te/(bat_te[0]*bat_te[1]))
    return te_loss, te_acc, output_te
```

```python
if run_type=='sweep1':
    print('[Training]%dch_%.1f_model%d'%(ch_num, time_len, model_num))
    torch.manual_seed(37)
    torch.cuda.manual_seed_all(37)
    torch.backends.cudnn.deterministic = True
    
    testX = np.load('../../../npy_data/sim_data%d/%dch_2/%dch_2_%.1fms_testX.npy'\
                    %(data_num, ch_num, ch_num, time_len))
    testY = np.load('../../../npy_data/sim_data%d/%dch_2/%dch_2_%.1fms_testY.npy'\
                    %(data_num, ch_num, ch_num, time_len))

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    batch_num = int(len(testX))
    batch_te = [int(len(testX)/batch_num), batch_num]
    
    testX = torch.Tensor(testX).to(device)
    testY = torch.Tensor(testY).to(device)
    print('%dch data shape  -  %.1f ms'%(ch_num, time_len))
    print('test set :', np.shape(testX) ,np.shape(testY))
    
    loss_func=nn.BCELoss()
    model = nn.DataParallel(Model(loss_func)).to(device)

    model.load_state_dict(torch.load('../../../ckpt/model%d/sim_data%d/19ch_2/%.1f_ckpt_1000.pt'\
                                     %(model_num, data_num, time_len)))

    test_loss = []
    test_acc = []
    test_loss, test_acc, _ = test(model, testX, testY, batch_te, device, 
                               test_loss, test_acc)
    print(' test acc : %.7f'%test_acc[-1])
    ##############################################
    ##############################################
    
    testX_ = np.load('../../../npy_data/sim_data%d_2/%dch_1/%dch_1_%.1fms_testX.npy'\
                    %(data_num, ch_num, ch_num, time_len))
    testY_ = np.load('../../../npy_data/sim_data%d_2/%dch_1/%dch_1_%.1fms_testY.npy'\
                    %(data_num, ch_num, ch_num, time_len))

    batch_num_ = int(len(testX_))
    batch_te_ = [int(len(testX_)/batch_num_), batch_num_]
    
    testX_ = torch.Tensor(testX_).to(device)
    testY_ = torch.Tensor(testY_).to(device)
    print('%dch data shape  -  %.1f ms'%(ch_num, time_len))
    print('test set :', np.shape(testX_) ,np.shape(testY_))

    test_loss_ = []
    test_acc_ = []
    test_loss_, test_acc_, output_te_ = test(model, testX_, testY_, batch_te_, device, 
                               test_loss_, test_acc_)
    print(' test acc : %.7f'%test_acc_[-1])
    
    inference_map = np.zeros((20,20))
    output_te_ = output_te_.cpu().detach().numpy()
    for k1 in range(20):
        for k2 in range(100):
            for k3 in output_te_[k1*100+k2]:
                inference_map[k1,k3] += 1

    plt.imshow(inference_map)
    plt.xlabel('ground truth')
    plt.ylabel('inference')
    plt.show()
```
