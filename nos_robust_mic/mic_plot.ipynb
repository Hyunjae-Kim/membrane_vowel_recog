```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#################mic fft plot

for SNR in [15,0,-15]:
    mic_teX = np.load('npy_data/mic/%ddb/%ddb_mic_20.0ms_testX.npy'%(SNR,SNR))
    print(np.shape(mic_teX))
    for k in [0,10,18]:
        data = mic_teX[0][k]
        fft_data = np.abs(np.fft.rfft(data))
        
        df = 1/0.02
        freq_axis = np.arange(len(fft_data))*df
        plt.plot(freq_axis, fft_data)
        plt.xlabel('Frequency (Hz)')
        plt.title('channel : %d   //  SNR : %d dB'%(k+1, SNR))
        plt.savefig('img/mic_fft_plot/%dch_%ddB_fft.png'%(k+1, SNR))
        plt.show()
```

```python
#################mic raw plot

for k in [0,10,18]:
    for SNR in [15,0,-15]:
        mic_teX = np.load('npy_data/mic/%ddb/%ddb_mic_20.0ms_testX.npy'%(SNR,SNR))
        print(np.shape(mic_teX))
        
        data = mic_teX[0][k]
        time_axis = np.arange(len(data))*0.1
        plt.plot(time_axis, data, label = '%d dB'%SNR)
        plt.title('%d channel'%(k+1))
        plt.xlabel('Time (ms)')
        plt.legend()
        
    plt.savefig('img/mic_raw_plot/%d ch.png'%(k+1))
    plt.show()
        
```
