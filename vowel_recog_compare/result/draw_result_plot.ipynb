```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_num = 3
data_list = ['mem', 'raw', 'fft', 'mfcc']
time_list = [20.0, 10.0, 5.0, 1.0, 0.5]
```

```python
for k1 in range(len(data_list)):
    for k2 in range(len(time_list)):
        file_tr = np.loadtxt('model%d/%s/%.1f_acc_tr.txt'%(model_num, data_list[k1], time_list[k2]))
        file_te = np.loadtxt('model%d/%s/%.1f_acc_te.txt'%(model_num, data_list[k1], time_list[k2]))
        
        plt.plot(file_tr, 'k', label = 'train')
        plt.plot(file_te, 'r', label = 'test')
        plt.xlabel('iteration')
        plt.xlabel('accuracy')
        plt.ylim((0.2,1.05))
        plt.legend()
        plt.savefig('../img/model%d/%s/%.1f_acc_tr.png'%(model_num, data_list[k1], time_list[k2]))
        plt.show()
        
```
