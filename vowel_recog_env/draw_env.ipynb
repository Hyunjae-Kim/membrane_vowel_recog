```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert
```

```python
def extract_env(wav):
    hilb_wav = hilbert(wav)
    amp_envelope = np.abs(hilb_wav)
    hilb_phase = np.unwrap(np.angle(hilb_wav))
    temp_fine = np.cos(hilb_phase)
    return amp_envelope, temp_fine
```

```python
data = np.load('npy_data/sim_data2/raw/raw_20.0ms_testX.npy')
print(np.shape(data))

data = np.load('npy_data/sim_data2/9ch/9ch_20.0ms_testX.npy')
data = data[:,5,:]
print(np.shape(data))
```

```python
count = -1
for k1 in range(20):
    for k2 in range(k1+1,20):
        print(count+1, 600 + k1*5, 600 + k2*5)
        count += 1
        if count==18:
            raw = data[count*100]
            env, temp = extract_env(raw)

            plt.plot(raw)
            plt.plot(env)
            plt.savefig('img/raw_env.png')
            plt.show()
            
            plt.plot(temp)
            plt.savefig('img/tfs.png')
            plt.show()

            plt.plot(temp*env,'r')
            plt.plot(raw, 'k')
            plt.savefig('img/restore.png')
            plt.show()
            break
    break
        
```
